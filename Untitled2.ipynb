```python
import pandas as pd
import numpy as np
from PIL import Image
import zipfile
from io import BytesIO
import torch
from torch.utils.data import Dataset, DataLoader, random_split
import os
```

```python
class LookLabDataset(Dataset):
    def __init__(self, csv_file, zip_path):
        self.data = pd.read_csv(csv_file)
        self.zip_path = zip_path
        self.zipfile = zipfile.ZipFile(zip_path, 'r')
        self.label_map = {label: idx for idx, label in enumerate(sorted(self.data['style'].unique()))}
        self.inv_label_map = {v: k for k, v in self.label_map.items()}

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        img_name = row['filename']
        img_data = self.zipfile.read(f"images/{img_name}")
        image = Image.open(BytesIO(img_data)).convert("RGB")
        image = image.resize((128, 128))
        image = np.array(image) / 255.0
        image = (image - 0.5) / 0.5
        image = torch.tensor(image, dtype=torch.float32).permute(2, 0, 1)
        label = self.label_map[row['style']]
        return image, label
```

```python
dataset = LookLabDataset(csv_file=labels_csv, zip_path=images_zip)

train_size = int(0.7 * len(dataset))
val_size = len(dataset) - train_size

train_dataset, val_dataset = random_split(dataset, [train_size, val_size], generator=torch.Generator().manual_seed(42))

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

print(f"‚úÖ Loaded: {train_size} training + {val_size} validation samples")
```

```python
# Usage-based categories (prioritized if not 'Casual')
usage_to_style = {
    'Smart Casual': 'Classic',
    'Formal': 'Classic',
    'Party': 'Glam',
    'Sports': 'Minimalist',
    'Travel': 'Boho',
    'Ethnic': 'Boho',
    'Home': 'Minimalist'
}

# Keyword-based mappings
keyword_map = {
    'Glam': ['glitter', 'sequin', 'velvet', 'evening', 'cocktail', 'shine', 'sparkle'],
    'Old Money': ['polo', 'blazer', 'cardigan', 'collar', 'pleated', 'buttondown', 'stripe'],
    'Grunge': ['graphic', 'ripped', 'distressed', 'flannel', 'hoodie', 'denim', 'washed'],
    'Classic': ['formal', 'trouser', 'chino', 'pant', 'white', 'navy', 'neutral'],
    'Minimalist': ['solid', 'plain', 'basic', 'monochrome', 'structured', 'minimal'],
    'Boho': ['kurta', 'ethnic', 'tunic', 'embroidery', 'anarkali', 'flare', 'kaftan'],
    'Romantic': ['lace', 'ruffle', 'pastel', 'bow', 'fit & flare', 'frill', 'pink'],
    'Casual': ['t-shirt', 'tee', 'tank', 'top', 'jogger', 'sweatpant']
}

# Footwear terms
footwear_keywords = ['shoe', 'shoes', 'sandal', 'sandals', 'slipper', 'slippers', 'heels', 'flats', 'flip flops', 'loafer', 'boots']

# Labeling function
def infer_style(usage, product_name):
    product_name = str(product_name).lower() if pd.notnull(product_name) else ''

    if pd.notnull(usage) and usage != "Casual":
        if usage in usage_to_style:
            return usage_to_style[usage]

    if any(word in product_name for word in footwear_keywords):
        return "Footwear"

    if "shirt" in product_name:
        if any(x in product_name for x in ["polo", "striped", "check", "checked"]):
            return "Old Money"
        elif "graphic" in product_name:
            return "Grunge"
        elif any(x in product_name for x in ["solid", "plain"]):
            return "Minimalist"
        elif any(x in product_name for x in ["floral", "lace", "pink"]):
            return "Romantic"
        elif any(x in product_name for x in ["embroidered", "printed", "kurta"]):
            return "Boho"
        else:
            return "Casual"

    for style, keywords in keyword_map.items():
        if any(keyword in product_name for keyword in keywords):
            return style

    return "Unique"
# Apply logic
styles_df['style'] = styles_df.apply(lambda row: infer_style(row['usage'], row['productDisplayName']), axis=1)

# Keep only required columns
labels_df = styles_df[['filename', 'style']].dropna()

# Save final labeled file
labels_out_path = os.path.join(shared_folder, "labels_final.csv")
labels_df.to_csv(labels_out_path, index=False)

print(f"labels_final.csv created with {len(labels_df)} labeled rows")
```

```python
import torch.nn as nn

class LookLabCNN(nn.Module):
    def __init__(self, num_classes):
        super(LookLabCNN, self).__init__()

        self.conv_layers = nn.Sequential(
            nn.Conv2d(3, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )

        self.fc_layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(16 * 16 * 128, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        x = self.conv_layers(x)
        x = self.fc_layers(x)
        return x
```

```python
import torch.optim as optim

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("üöÄ Using device:", device)

model = LookLabCNN(num_classes=len(dataset.label_map)).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
train_accuracies = []
val_accuracies = []

num_epochs = 5  # Or however many you want

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    train_acc = 100 * correct / total
    train_accuracies.append(train_acc)

    # Validation
    model.eval()
    val_correct = 0
    val_total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            val_total += labels.size(0)
            val_correct += (predicted == labels).sum().item()

    val_acc = 100 * val_correct / val_total
    val_accuracies.append(val_acc)

    print(f"Epoch {epoch+1}/{num_epochs} - Train Acc: {train_acc:.2f}% | Val Acc: {val_acc:.2f}%")
```

```python
import matplotlib.pyplot as plt

epochs = range(1, num_epochs + 1)

plt.figure(figsize=(8, 5))
plt.plot(epochs, train_accuracies, label='Training Accuracy')
plt.plot(epochs, val_accuracies, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('Training vs Validation Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
